### 1.3 æ¢ä»¶åˆ¤æ–·èˆ‡è¿´åœˆ

```python
# if-elif-else æ¢ä»¶åˆ¤æ–·
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"åˆ†æ•¸: {score}, ç­‰ç´š: {grade}")

# ä¸‰å…ƒé‹ç®—å­
result = "åŠæ ¼" if score >= 60 else "ä¸åŠæ ¼"
print(f"ä¸‰å…ƒé‹ç®—: {result}")
```

```python
# for è¿´åœˆ + enumerate
fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits):
    print(f"  [{i}] {fruit}")

# while è¿´åœˆ
print("\nwhile è¿´åœˆå€’æ•¸:")
count = 3
while count > 0:
    print(f"  {count}...")
    count -= 1
print("  ğŸš€ ç™¼å°„ï¼")

# break & continue
print("\nbreak & continue ç¤ºç¯„:")
for i in range(10):
    if i == 3:
        continue  # è·³é 3
    if i == 7:
        break     # åˆ° 7 åœæ­¢
    print(f"  i = {i}")
```

### 1.4 å‡½å¼èˆ‡ Lambda

```python
# å®šç¾©å‡½å¼
def calculate_bmi(weight, height_cm):
    """è¨ˆç®— BMI å€¼"""
    height_m = height_cm / 100
    return round(weight / (height_m ** 2), 1)

bmi = calculate_bmi(70, 175)
print(f"BMI = {bmi}")

# é è¨­åƒæ•¸å€¼
def greet(name, greeting="ä½ å¥½"):
    return f"{greeting}, {name}!"

print(greet("Alice"))
print(greet("Bob", "å—¨"))

# Lambda åŒ¿åå‡½å¼
square = lambda x: x ** 2
print(f"\nlambda square(5) = {square(5)}")

# æ­é… map ä½¿ç”¨
numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(f"map + lambda: {numbers} -> {squared}")
```

### 1.5 ä¸²åˆ—æ¨å°å¼ï¼ˆList Comprehensionï¼‰

```python
# åŸºæœ¬æ¨å°å¼
squares = [x**2 for x in range(1, 11)]
print(f"å¹³æ–¹æ•¸: {squares}")

# æ¢ä»¶ç¯©é¸
evens = [x for x in range(1, 21) if x % 2 == 0]
print(f"å¶æ•¸:   {evens}")

# å·¢ç‹€æ¨å°å¼
matrix = [[i * j for j in range(1, 4)] for i in range(1, 4)]
print(f"çŸ©é™£:   {matrix}")

# å­—å…¸æ¨å°å¼
word = "hello"
char_count = {ch: word.count(ch) for ch in set(word)}
print(f"å­—å…ƒè¨ˆæ•¸: {char_count}")
```

## 3. Pandas è³‡æ–™è®€å–èˆ‡æ¢ç´¢


### 3.1 å»ºç«‹æ¨¡æ“¬è³‡æ–™é›†

> æˆ‘å€‘å»ºç«‹ä¸€ä»½åŒ…å«å„é¡å¸¸è¦‹è³‡æ–™å“è³ªå•é¡Œçš„æ¨¡æ“¬è³‡æ–™é›†ï¼Œç”¨æ–¼å¾ŒçºŒæ‰€æœ‰å¯¦ä½œç·´ç¿’ã€‚


```python
import pandas as pd
np.random.seed(42)

n = 200

# ç”¢ç”Ÿæ¨¡æ“¬è³‡æ–™ï¼ˆæ•…æ„åŠ å…¥å„é¡å“è³ªå•é¡Œï¼‰
data = {
    "student_id": list(range(1, n + 1)),
    "name": [f"Student_{i}" for i in range(1, n + 1)],
    "gender": np.random.choice(["M", "F", "male", "female", " F", "M "], n),  # æ ¼å¼ä¸ä¸€è‡´
    "age": np.random.choice(
        list(range(18, 30)) + [None, None, -5, 150],  # å«ç¼ºå¤±å€¼èˆ‡ç•°å¸¸å€¼
        n
    ),
    "department": np.random.choice(
        ["CS", "EE", "ME", "cs", "Cs", None],  # å¤§å°å¯«ä¸ä¸€è‡´ + ç¼ºå¤±
        n
    ),
    "math_score": np.where(
        np.random.random(n) > 0.1,
        np.random.normal(75, 15, n).round(1),
        np.nan  # ç´„ 10% ç¼ºå¤±
    ),
    "english_score": np.where(
        np.random.random(n) > 0.08,
        np.random.normal(70, 12, n).round(1),
        np.nan  # ç´„ 8% ç¼ºå¤±
    ),
    "enrollment_date": pd.date_range("2022-09-01", periods=n, freq="h").astype(str),
    "email": [
        f"student_{i}@school.edu" if np.random.random() > 0.05
        else f"  Student_{i}@SCHOOL.EDU  "  # å«ç©ºç™½èˆ‡å¤§å°å¯«å•é¡Œ
        for i in range(1, n + 1)
    ],
}

df = pd.DataFrame(data)

# æ•…æ„åŠ å…¥é‡è¤‡åˆ—
dup_rows = df.sample(10, random_state=42)
df = pd.concat([df, dup_rows], ignore_index=True)

# æ‰“äº‚é †åº
df = df.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"âœ… æ¨¡æ“¬è³‡æ–™é›†å·²å»ºç«‹ï¼Œå…± {len(df)} ç­† x {df.shape[1]} æ¬„")
df.head(10)
```

### 4.1 è™•ç†ç¼ºå¤±å€¼ï¼ˆMissing Valuesï¼‰

#### 4.1.1 åµæ¸¬ç¼ºå¤±å€¼

```python
# å„æ¬„ç¼ºå¤±å€¼æ•¸é‡èˆ‡æ¯”ä¾‹
missing_stats = pd.DataFrame({
    'ç¼ºå¤±æ•¸é‡': df_clean.isnull().sum(),
    'ç¼ºå¤±æ¯”ä¾‹(%)': (df_clean.isnull().mean() * 100).round(2)
})
missing_stats = missing_stats[missing_stats['ç¼ºå¤±æ•¸é‡'] > 0].sort_values('ç¼ºå¤±æ¯”ä¾‹(%)', ascending=False)
print("ğŸ“‹ ç¼ºå¤±å€¼çµ±è¨ˆ:")
display(missing_stats)
```

```python
# è¦–è¦ºåŒ–ç¼ºå¤±å€¼åˆ†å¸ƒ
import missingno as msno

fig, ax = plt.subplots(figsize=(10, 4))
msno.bar(df_clean, ax=ax, color="steelblue", fontsize=10)
plt.title("å„æ¬„ä½è³‡æ–™å®Œæ•´åº¦", fontsize=14)
plt.tight_layout()
plt.show()
```

#### 4.1.2 å¡«è£œèˆ‡åˆªé™¤ç¼ºå¤±å€¼

```python
# --- æ•¸å€¼æ¬„ä½ï¼šç”¨ä¸­ä½æ•¸å¡«è£œ ---
print("æ•¸å€¼æ¬„ä½ç¼ºå¤±å€¼å¡«è£œï¼ˆä¸­ä½æ•¸ï¼‰:")
for col in ['math_score', 'english_score']:
    median_val = df_clean[col].median()
    missing_count = df_clean[col].isnull().sum()
    df_clean[col].fillna(median_val, inplace=True)
    print(f"  {col}: å¡«è£œ {missing_count} ç­†ï¼Œä¸­ä½æ•¸ = {median_val}")

# --- é¡åˆ¥æ¬„ä½ï¼šç”¨çœ¾æ•¸å¡«è£œ ---
print("\né¡åˆ¥æ¬„ä½ç¼ºå¤±å€¼å¡«è£œï¼ˆçœ¾æ•¸ï¼‰:")
for col in ['department', 'gender']:
    mode_val = df_clean[col].mode()[0]
    missing_count = df_clean[col].isnull().sum()
    df_clean[col].fillna(mode_val, inplace=True)
    print(f"  {col}: å¡«è£œ {missing_count} ç­†ï¼Œçœ¾æ•¸ = '{mode_val}'")

# --- age æ¬„ä½ï¼šå…ˆè½‰æ•¸å€¼å†å¡«è£œ ---
df_clean['age'] = pd.to_numeric(df_clean['age'], errors='coerce')
age_median = df_clean['age'].median()
age_missing = df_clean['age'].isnull().sum()
df_clean['age'].fillna(age_median, inplace=True)
print(f"\n  age: å¡«è£œ {age_missing} ç­†ï¼Œä¸­ä½æ•¸ = {age_median}")

# ç¢ºèª
print(f"\nâœ… å‰©é¤˜ç¼ºå¤±å€¼: {df_clean.isnull().sum().sum()}")
```

### 4.2 è™•ç†é‡è¤‡è³‡æ–™ï¼ˆDuplicatesï¼‰

```python
# åµæ¸¬é‡è¤‡
dup_count = df_clean.duplicated().sum()
print(f"ğŸ” å®Œå…¨é‡è¤‡çš„åˆ—æ•¸: {dup_count}")

# æŸ¥çœ‹é‡è¤‡çš„åˆ—
if dup_count > 0:
    print("\né‡è¤‡åˆ—ç¯„ä¾‹:")
    display(df_clean[df_clean.duplicated(keep=False)].sort_values('student_id').head(6))

# ç§»é™¤é‡è¤‡ï¼ˆä¿ç•™ç¬¬ä¸€ç­†ï¼‰
before = len(df_clean)
df_clean.drop_duplicates(keep='first', inplace=True)
df_clean.reset_index(drop=True, inplace=True)
after = len(df_clean)
print(f"\nâœ… ç§»é™¤é‡è¤‡: {before} -> {after} åˆ—ï¼ˆåˆªé™¤ {before - after} ç­†ï¼‰")
```

```python
# ä¹Ÿå¯ä»¥ä¾ç‰¹å®šæ¬„ä½åˆ¤æ–·é‡è¤‡
dup_by_id = df_clean.duplicated(subset=['student_id']).sum()
print(f"ä¾ student_id æª¢æŸ¥é‡è¤‡: {dup_by_id} ç­†")

# è‹¥æœ‰ student_id é‡è¤‡ï¼Œåªä¿ç•™ç¬¬ä¸€ç­†
if dup_by_id > 0:
    df_clean.drop_duplicates(subset=['student_id'], keep='first', inplace=True)
    df_clean.reset_index(drop=True, inplace=True)
    print(f"âœ… è™•ç†å¾Œ: {len(df_clean)} ç­†")
```

#### 4.4.2 IQR æ³•åµæ¸¬èˆ‡è™•ç†

```python
def detect_outliers_iqr(series, factor=1.5):
    """ä½¿ç”¨ IQR æ³•åµæ¸¬ç•°å¸¸å€¼"""
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - factor * IQR
    upper = Q3 + factor * IQR
    outliers = (series < lower) | (series > upper)
    return outliers, lower, upper

# åµæ¸¬å„æ•¸å€¼æ¬„ä½çš„ç•°å¸¸å€¼
print("ğŸ“‹ IQR æ³•ç•°å¸¸å€¼åµæ¸¬:")
print(f"{'æ¬„ä½':<18} {'ä¸‹ç•Œ':>8} {'ä¸Šç•Œ':>8} {'ç•°å¸¸ç­†æ•¸':>8}")
print("-" * 46)

for col in ['age', 'math_score', 'english_score']:
    mask, lower, upper = detect_outliers_iqr(df_clean[col])
    count = mask.sum()
    print(f"{col:<18} {lower:>8.1f} {upper:>8.1f} {count:>8}")
```

```python
# è™•ç† age çš„ç•°å¸¸å€¼ï¼ˆæˆªæ–·æ³• / Cappingï¼‰
mask, lower, upper = detect_outliers_iqr(df_clean['age'])
before_outliers = mask.sum()

# æ–¹æ³•ä¸€ï¼šæˆªæ–·ï¼ˆWinsorizingï¼‰â€” å°‡è¶…å‡ºç¯„åœçš„å€¼è¨­ç‚ºé‚Šç•Œå€¼
df_clean.loc[df_clean['age'] < lower, 'age'] = lower
df_clean.loc[df_clean['age'] > upper, 'age'] = upper

# ä¹Ÿè™•ç†æ˜é¡¯ä¸åˆç†çš„å€¼
df_clean.loc[df_clean['age'] < 15, 'age'] = 15
df_clean.loc[df_clean['age'] > 35, 'age'] = 35

print(f"âœ… age ç•°å¸¸å€¼è™•ç†: {before_outliers} ç­† -> æˆªæ–·è‡³åˆç†ç¯„åœ")
print(f"   age ç¯„åœ: {df_clean['age'].min():.0f} ~ {df_clean['age'].max():.0f}")
```

### 7.2 å°‡æ¸…æ´—æ­¥é©Ÿå°è£ç‚ºè‡ªå‹•åŒ–å‡½å¼

```python
def clean_student_data(df_input):
    """
    å®Œæ•´çš„å­¸ç”Ÿè³‡æ–™æ¸…æ´— Pipeline
    è¼¸å…¥: åŸå§‹ DataFrame
    è¼¸å‡º: æ¸…æ´—å¾Œçš„ DataFrame
    """
    df = df_input.copy()

    # Step 1: å‹æ…‹è½‰æ›
    df['age'] = pd.to_numeric(df['age'], errors='coerce')
    df['enrollment_date'] = pd.to_datetime(df['enrollment_date'], errors='coerce')

    # Step 2: ç¼ºå¤±å€¼å¡«è£œ
    df['math_score'].fillna(df['math_score'].median(), inplace=True)
    df['english_score'].fillna(df['english_score'].median(), inplace=True)
    df['age'].fillna(df['age'].median(), inplace=True)
    df['department'].fillna(df['department'].mode()[0], inplace=True)
    df['gender'].fillna(df['gender'].mode()[0], inplace=True)

    # Step 3: ç§»é™¤é‡è¤‡
    df.drop_duplicates(keep='first', inplace=True)
    df.drop_duplicates(subset=['student_id'], keep='first', inplace=True)

    # Step 4: å­—ä¸²æ¸…æ´—
    df['gender'] = (df['gender'].astype(str).str.strip().str.lower()
                    .replace({'male': 'm', 'female': 'f'})
                    .map({'m': 'M', 'f': 'F'}))
    df['department'] = df['department'].astype(str).str.strip().str.upper()
    df['email'] = df['email'].str.strip().str.lower()

    # Step 5: ç•°å¸¸å€¼æˆªæ–·
    df.loc[df['age'] < 15, 'age'] = 15
    df.loc[df['age'] > 35, 'age'] = 35

    # Step 6: ç‰¹å¾µå·¥ç¨‹
    df['total_score'] = df['math_score'] + df['english_score']
    df['avg_score'] = df['total_score'] / 2

    df.reset_index(drop=True, inplace=True)
    return df

# æ¸¬è©¦è‡ªå‹•åŒ–å‡½å¼
df_auto = clean_student_data(df_raw)
print(f"âœ… è‡ªå‹•åŒ–æ¸…æ´—å®Œæˆ: {df_auto.shape}")
print(f"   ç¼ºå¤±å€¼: {df_auto.isnull().sum().sum()}")
print(f"   é‡è¤‡åˆ—: {df_auto.duplicated().sum()}")
```
